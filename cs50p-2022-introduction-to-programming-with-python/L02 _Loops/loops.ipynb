### Notes

- Loops, its the ability in Python and a lot of other programming languages to do something again 
and again thus repeating a set of instructions multiple times until a certain condition is met.
- A `while` loop is a control flow statement that repeatedly executes a block of code as long as a 
specified condition remains `True`.

```python
# Print "meow" three times
print('meow')
print('meow')
print('meow')
```

```python
# Print "meow" three times using a while loop
i = 3

while i != 0:
    print('meow')
    i = i - 1
```

```python
# Another way to print "meow" three times using a while loop
i = 0

while i < 3:
    print('meow')
    i += 1
```

### Notes

- A `for` loop is a control flow statement used to iterate over a sequence (like a list, string,
tuple, or range) and execute a block of code once for each item in that sequence.
- A `list` is a built-in data type used to store an ordered, changeable (mutable) collection of
items. Each item can be of any data type, and items are enclosed in square brackets [].
- A `dict` (dictionary) is a built-in data type that stores data in keyâ€“value pairs, where each key 
is unique and used to access its corresponding value.

```python
# Print "meow" three times using a for loop
for i in [0, 1, 2]:
    print('meow')
```

```python
# Another way to print "meow" three times using a for loop
# range() function returns a sequence of numbers from 0 to 2
for _ in range(3):
    print('meow')
```

```python
# Print "meow" three times using string multiplication
print('meow\n' * 3, end='')
```

```python
# Prompt user for a positive integer n and print "meow" n times
# This while is an important paradigm called input validation
# It ensures that the user enters a valid input
while True:
    n = int(input("What's n? "))    # 3

    if n > 0:
        break

for _ in range(n):
    print('meow')
```

```python
# Refactored version using functions to prompt user and print "meow" n times
def main():
    number = get_number()
    meow(number)

def get_number():
    while True:
        n = int(input("What's n? "))    # 3

        if n > 0:
            return n

def meow(n):
    for _ in range(n):
        print('meow')

main()
```

```python
# Print each student's name individually
students = ['Harry', 'Hermione', 'Ron']

print(students[0])
print(students[1])
print(students[2])
```

```python
# Print each student's name using a for loop
students = ['Harry', 'Hermione', 'Ron']

for student in students:
    print(student)
```

```python
# Print each student's name using a for loop with index
students = ['Harry', 'Hermione', 'Ron']

for i in range(len(students)):
    print(i + 1, students[i])
```

```python
# Print each student's house from the dictionary individually
students = {
    'Harry': 'Gryffindor',
    'Hermione': 'Gryffindor',
    'Ron': 'Gryffindor',
    'Draco': 'Slytherin'
}

print(students['Harry'])
print(students['Hermione'])
print(students['Ron'])
print(students['Draco'])
```

```python
# Print each student's name along with their house using a for loop
students = {
    'Harry': 'Gryffindor',
    'Hermione': 'Gryffindor',
    'Ron': 'Gryffindor',
    'Draco': 'Slytherin',
}

for student in students:
    print(student, students[student], sep=', ')
```

```python
# Print each student's details from a list of dictionaries
students = [
    {'name': 'Harry', 'house': 'Gryffindor', 'patronus': 'Stag'},
    {'name': 'Hermione', 'house': 'Gryffindor', 'patronus': 'Otter'},
    {'name': 'Ron', 'house': 'Gryffindor', 'patronus': 'Jack Russell terrier'},
    {'name': 'Draco', 'house': 'Slytherin', 'patronus': None},
]

for student in students:
    print(student['name'], student['house'], student['patronus'], sep=', ')
```

```python
# Print a column of hashes of a given height
def main():
    print_column(3)

def print_column(height):
    print('#\n' * height, end='')

main()
```

```python
# Print a row of question marks of a given width
def main():
    print_row(4)

def print_row(width):
    print('?' * width)


main()
```

```python
# Print a square of size n x n using nested loops
def main():
    print_square(3)

def print_square(size):
    # For each row in square
    for i in range(size):

        # For each brick in row
        for j in range(size):

            # Print a brick without moving to the next line
            print('#', end='')

        print()  # Move to the next line after each row

main()
```

```python
# Refactored version of printing a square using helper functions
def main():
    print_square(3)

def print_square(size):
    for _ in range(size):
        print_row(size)

def print_row(width):
    print('#' * width)

main()
```

```python
# Final refactored version of printing a square without a helper function
def main():
    print_square(3)

def print_square(size):
    for _ in range(size):
        print("#" * size)

main()
```

### Notes

- In Python we have two type of loops
  - `for` loop: Iterates over sequences
  - `while` loop: Repeats code while a condition is `True`
- Python does not have `do-while` loop


```python
# A do-while loop runs at least once, then checks the condition
# We can mimic a do-while loop in Python using a while True loop with a break condition
count = 1

while True:
    print(count)
    count += 1

    if count > 5:
        break
```
