# CS50P - Lecture 1 - Conditionals

### Task

Write a program to demonstrate the use of conditional statements

### Notes

- Conditionals, or conditional statements, in Python and in other languages are this ability to 
ask questions and answer those questions, in order to decide if you want to execute a line of code
or some other line of code instead


```python
# Version 1: Basic version with separate if statements
# Disadvantage: All conditions are checked even if one is already true
x = int(input("What's x? "))    # 1
y = int(input("What's y? "))    # 2

# x < y is a boolean expression, which evaluates to True or False
if x < y:
    print('x is less than y')

if x > y:
    print('x is greater than y')

if x == y:
    print('x is equal to y')
```

```python
# Version 2: Improvised version with elif
x = int(input("What's x? "))    # 2
y = int(input("What's y? "))    # 1

if x < y:
    print('x is less than y')
elif x > y:
    print('x is greater than y')
elif x == y:
    print('x is equal to y')
```

```python
# Version 3: Improvised version with elif and else
x = int(input("What's x? "))  # 2
y = int(input("What's y? "))  # 2

if x < y:
    print("x is less than y")
elif x > y:
    print("x is greater than y")
else:
    print("x is equal to y")
```

```python
# Version 4: Using or operator
x = int(input("What's x? "))  # 1
y = int(input("What's y? "))  # 1

if x < y or x > y:
    print('x is not equal to y')
else:
    print('x is equal to y')
```

```python
# Version 5: Improvised version with not operator
x = int(input("What's x? "))  # 3
y = int(input("What's y? "))  # 3

if x != y:
    print('x is not equal to y')
else:
    print('x is equal to y')
```

```python
# Version 1: Simple grading system example
score = int(input('Score: '))  # 95

if score >= 90 and score <= 100:
    print('Grade: A')
elif score >= 80 and score < 90:
    print('Grade: B')
elif score >= 70 and score < 80:
    print('Grade: C')
elif score >= 60 and score < 70:
    print('Grade: D')
else:
    print('Grade: F')
```

```python
# Version 2: Improved grading system example
score = int(input('Score: '))  # 85

if 90 <= score <= 100:
    print('Grade: A')
elif 80 <= score < 90:
    print('Grade: B')
elif 70 <= score < 80:
    print('Grade: C')
elif 60 <= score < 70:
    print('Grade: D')
else:
    print('Grade: F')
```

```python
# Version 3: Simplified grading system example
score = int(input('Score: '))  # 75

if score >= 90:
    print('Grade: A')
elif score >= 80:
    print('Grade: B')
elif score >= 70:
    print('Grade: C')
elif score >= 60:
    print('Grade: D')
else:
    print('Grade: F')
```

```python
# Version 1: Simple even or odd evaluation example
x = int(input("What's x? "))    # 1

if x % 2 == 0:
    print('Even')
else:
    print('Odd')
```

```python
# Version 2: Using a function to check even or odd
def main():
    x = int(input("What's x? "))  # 2
    
    if is_even(x):
        print('Even')
    else:
        print('Odd')

def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False

main()
```

```python
# Version 3: Simplified function to check even or odd
def main():
    x = int(input("What's x? "))  # 3
    
    if is_even(x):
        print('Even')
    else:
        print('Odd')

def is_even(n):
    return True if n % 2 == 0 else False

main()
```

```python
# Version 4: Further simplified function to check even or odd
def main():
    x = int(input("What's x? "))  # 4

    if is_even(x):
        print('Even')
    else:
        print('Odd')

def is_even(n):
    return n % 2 == 0

main()
```

```python
# Version 1: House assignment using if-elif-else
name = input("What's your name? ")  # Hermione

if name == 'Harry' or name == 'Hermione' or name == 'Ron':
    print('Gryffindor')
elif name == 'Draco':
    print('Slytherin')
else:
    print('Who?')
```

```python
# Version 2: House assignment using match-case
name = input("What's your name? ")  # Draco

match name:
    case 'Harry' | 'Hermione' | 'Ron':
        print('Gryffindor')
    case 'Draco':
        print('Slytherin')
    case _:
        print('Who?')
```
